```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('DailyDelhiClimateTest.csv')
df.head()
```

```python
df=df.iloc[:,0:2]
df.head()
```

```python
plt.scatter(df['date'],df['meantemp'],color='red')
plt.plot(df['date'],df['meantemp'],color='black')
```

```python
x1=df['meantemp'].ewm(alpha=0.9).mean()
```

```python
x1
```

```python
df['ewma']=x1
df
```

```python
plt.scatter(df['date'],df['ewma'],color='red')
plt.plot(df['date'],df['ewma'],color='black')
```
