```python
import os 
import caer
import canaro
import numpy as np
import cv2 as cv
import gc
```

```python
##all the image should be of same size
IMG_SIZE=(80,80)
channels=1# since we donot require color in image
char_path=r'/kaggle/input/the-simpsons-characters-dataset'
```

```python
char_dict={}
for char in os.listdir(char_path):
    char_dict[char]=len(os.listdir(os.path.join(char_path,char)))
#here we are going to the path and grabing the name of the folder and calculating the number of image inside the folder and storing that information in the dictionary
 
#sort in descending order
char_dict=caer.sort_dict(char_dict,descending=True)
char_dict
```

```python
characters=[]
count=0
for i in char_dict:
    characters.append(i[0])
    count+=1
    if count>=10:
        break
characters
        
```

```python
#create the training data
train=caer.preprocessing_from_dir(char_path,characters,channels,channels=channels,IMG_SIZE=IMG_SIZE,isShuffle=True)
#it goes inside the directory and take all the images to the training process
```

```python
len(train)# no of elements in training dataset
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(30,30))
plt.imshow(train[0][0],cmap='gray')
plt.show()
```

```python
#separating the training set into features and labels
featureset,labels=caer.sep_train(train,IMG_SIZE=IMG_SIZE)
```

```python
#Normalize the featureset ==>(0,1)
featureset=caer.normalize(featureset)
# now we don't normalize the labels but instead we categorize the labels into binary format using encoding(0 and 1)
from tensorflow.keras.utils import to_categorical
labels=to_categorical(labels,len(characters))
```

```python
#creating trainging and validation data
x_train,x_val,y_train,y_val=caer.train_val_split(fatureset,labels,val_ration=.2)
```

```python
BATCH_SIZE=32
EPOCHS=10
```

```python
#Image data   generator
# it will generate new images from already existing images to help introduce some randomness to our network and make it perform better
datagen=canaro.generators.imageDataGenerator()
train_gen=datagen.flow(x_train,y_train,batch_size=BATCH_SIZE)
```

```python
# Creating the model
model=canaro.medels.createSimpsonsModels(IMG_SIZE=IMG_SIZE,channels=channels,output_dim=len(characters),loss='binary_crossentropy',decay=1e-6,learning_rate=0.001,momentum=0.9,nesterov=True)
```

```python
model.summary()
```

```python
# callbacks will schedule the learning rate
from tensorflow.keras.callbacks import LearningRateScheduler
callbacks_list=[LearningRateScheduler(canaro.lr_schedule)]
```

```python
training=model.fit(train_gen,
                  steps_per_epoch=len(x_train)//BATCH_SIZE,
                  epochs=EPOCHS,
                  validation_data=(x_val,y_val),
                  validation_steps=len(y_val)//BATCH_SIZE,
                  callbacks=callbacks_list)
```

```python
#now we are going to use the opencv to test how good the model is 
# we are going to pass the image at the perticular file path 

test_path=r''
img=cv.imread(test_path)
plt.imshow(img)
plt.show()
```

```python
def prepare(img):
    img=cv.cvtColor(img,cv.COLOR_BAYER_BGGR2BGR)
    img=cv.resize(img,IMG_SIZE)
    img=caer.reshape(img,IMG_SIZE,1)
    return img
```

```python
predictions=model.predict(prepare(img))
```
